```python
import pandas as pd

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

display(df_train.head())
display(df_test.head())
print(df_train.info())
print(df_test.info())
```

```python
for col in ['GrLivArea', 'BedroomAbvGr', 'FullBath']:
    missing_train = df_train[col].isnull().sum()
    missing_test = df_test[col].isnull().sum() if col in df_test.columns else 0 # Handle potential KeyError
    print(f"Missing values in {col}: Train - {missing_train}, Test - {missing_test}")

def treat_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df

for col in ['GrLivArea', 'BedroomAbvGr', 'FullBath']:
    df_train = treat_outliers(df_train, col)
    if col in df_test.columns:
        df_test = treat_outliers(df_test, col)

for col in ['GrLivArea', 'BedroomAbvGr', 'FullBath', 'SalePrice']:
    if col in df_train.columns:
        print(f"Data type of {col} in df_train: {df_train[col].dtype}")
        if df_train[col].dtype != 'int64' and df_train[col].dtype != 'float64':
             df_train[col] = pd.to_numeric(df_train[col], errors='coerce')
    if col in df_test.columns:
        print(f"Data type of {col} in df_test: {df_test[col].dtype}")
        if df_test[col].dtype != 'int64' and df_test[col].dtype != 'float64':
            df_test[col] = pd.to_numeric(df_test[col], errors='coerce')
```

```python
df_train['BedroomBathInteraction'] = df_train['BedroomAbvGr'] * df_train['FullBath']
df_test['BedroomBathInteraction'] = df_test['BedroomAbvGr'] * df_test['FullBath']

relevant_cols = ['GrLivArea', 'BedroomAbvGr', 'FullBath', 'SalePrice', 'BedroomBathInteraction']

for col in relevant_cols:
    if col in df_train.columns:
        if not pd.api.types.is_numeric_dtype(df_train[col]):
            print(f"Converting column '{col}' in df_train to numeric")
            df_train[col] = pd.to_numeric(df_train[col], errors='coerce')
    if col in df_test.columns:
        if not pd.api.types.is_numeric_dtype(df_test[col]):
            print(f"Converting column '{col}' in df_test to numeric")
            df_test[col] = pd.to_numeric(df_test[col], errors='coerce')
```
